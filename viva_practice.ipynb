```python
job_counts = sales_data['job_title'].value_counts().reset_index()
job_counts.columns=['job_title', 'job_count']

top_15_job= job_counts.sort_values(by='job_count',ascending=False).head(15)

top_15_job
```

```python
job_titles= top_15_job['job_title']
job_counts= top_15_job['job_count']

plt.figure(figsize=(15,8))
bars=plt.bar(job_titles, job_counts, color='skyblue')
plt.xlabel('Job Title')
plt.ylabel('Top 15 Job Titles by Count')
plt.xticks(rotation=90)

for i, v in enumerate(job_counts):
    plt.text(i, v+10, str(v), ha='center', va='bottom', fontsize=9)
plt.show()
```

```python
maxsalary_job= sales_data.groupby('job_title')[['salary_in_usd']].max()

highest_salaries= maxsalary_job.sort_values(by='salary_in_usd',ascending=False)
highest_salaries
```

```python
plt.figure(figsize=(25,12))

highest_salaries.plot(kind='bar', color='skyblue')

plt.title('job with their highest salary')

plt.xlabel('Job Title')
plt.ylabel('Salary USD')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
